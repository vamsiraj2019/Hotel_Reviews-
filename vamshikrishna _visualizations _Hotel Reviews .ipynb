# Problem Statement for Visualizations

- Develop visualizations and gain insights about the Sentiment whether excellent/good/bad from the data which contained customer   reviews about different hotels.
- The objective of the visualizations was to find any possible relationship between the features (Hotelid, userid, Date and reviewtext) and Sentiment.
- Such relations would help us gain insights into things such as -
 customer experience
- Understand seasonal demand of Hotels
- Check out the trend of the demand at Hotels Vs customer experience
- The above insights are helpful for both customers and Hotels -
- Customers
- Plan their vacation/travel in a better way by choosing Hotels with better reviews
- Plan their trips well ahead of time based on the seasonal demand of Hotels
  Hotels
- Forecast the demand across different seasons
- come up with different promotions to improve their occupancy levels and improve sales/profits
- Observing customer experience via reviews can help them to improve their operations efficiency
- Insights/Obervations
- Most of data comprised of good reviews (50%)
- August and September months & Q3 contributed to more reviews
- One Hotel 188 had 500+ reviews which is close to 8.4% of total data
- The number of words had a direct relation to the proportion of bad reviews

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color = 'white',
        max_words = 200,
        max_font_size = 40, 
        scale = 3,
        random_state = 42
    ).generate(str(data))

    fig = plt.figure(1, figsize = (20, 20))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize = 20)
        fig.subplots_adjust(top = 2.3)

    plt.imshow(wordcloud)
    plt.show()
    
# print wordcloud
show_wordcloud(hotel_Reviews["reviewtext"])
```
